```python
import sympy as sp
import numpy as np
import math
```

```python
# Definir la matriz
n = 5 # modulo
# numero de filas
filas = 3
# numero de columnas
columnas = 3

# Crear una matriz simbólica de tamaño filas x columnas
A = sp.Matrix(filas, columnas, lambda i, j: sp.symbols(f'a{i+1}{j+1}'))

# Rellenar la matriz con valores específicos

# lista de valores para rellenar la matriz
valores = [1, 2, 3,
           4, 5, 6,
           7, 8, 9]

# Rellenar la matriz con los valores
for i in range(filas):
    for j in range(columnas):
        A[i, j] = valores[i * columnas + j]
        
A = A.applyfunc(lambda x: x % n)

# Mostrar la matriz
display(A)
```

```python
# Calcular la matriz equivalente y equivalente reducida a A
E = A.rref()[0]
display(E)
```

```python
# Calcular matrices escalonadas y escalonadas reducidas en Zn

if n is not None:
    A_mod_n = A.applyfunc(lambda x: x % n)
    E_mod_n = A_mod_n.rref()[0]
    display(E_mod_n)
else:
    E = A.rref()[0]
    display(E)
```

```python
# Calcular la inversa de la matriz

if filas != columnas:
    print("La matriz no es cuadrada, por lo tanto no es invertible.")

else:
    det = A.det()

    if n is not None:
        if math.gcd(det, n) == 1:
            A_inv = A.inv_mod(n)
            display(A_inv)
        else:
            print("La matriz no es invertible módulo", n)
    else:
        if det != 0:
            A_inv = A.inv()
            display(A_inv)
        else:
            print("La matriz no es invertible.")
```

```python
# Calcular el rango de la matriz
if n is not None:
    A_mod_n = A.applyfunc(lambda x: x % n)
    rango = A_mod_n.rank()
    print(f"El rango de la matriz A módulo {n} es: {rango}")
else:
    rango = A.rank()
    print(f"El rango de la matriz A es: {rango}")
```

```python
# Calcular el determinante de la matriz si es posible

if filas == columnas:
    if (n is not None):
        det_A_mod_n = A.det() % n
        print(f"El determinante de la matriz A módulo {n} es: {det_A_mod_n}")
    else:
        det_A = A.det()
        display(det_A)
else:
    print("La matriz no es cuadrada, no se puede calcular el determinante.")
```
